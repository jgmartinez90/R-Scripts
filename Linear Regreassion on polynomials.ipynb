```r
#This script creates a model to fit on a polynomial scatter plot

# x-axis
x <- runif(300,  min=-10, max=10) 
#y-axis length of x = 300
y <- 0.1*x^3 - 0.5 * x^2 - x + 10 + rnorm(length(x),0,8) 
 
# plot of x and y
plot(x,y,col=rgb(0.4,0.4,0.8,0.6),pch=16 , cex=1.3) 
 
# Create a Model 
model <- lm(y ~ x + I(x^2) + I(x^3))
 
# Stats of the model 
summary(model)
print("Model and Coefficients")
model$coefficients

 
# model added in the plot
myPredict <- predict( model ) 
ix <- sort(x,index.return=T)$ix
lines(x[ix], myPredict[ix], col=2, lwd=2 )  

# Add model on the plot with rounded coefficients
coeff <- round(model$coefficients , 2)
text(3, -70 , paste("Model : ",coeff[1] , " + " , coeff[2] , "*x"  , "+" , coeff[3] , "*x^2" , "+" , coeff[4] , "*x^3" , "\n\n")) 
```
